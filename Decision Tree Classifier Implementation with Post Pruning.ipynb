```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
print(dataset.DESCR)
```

```python
# IMPORTING DATASET BY SEABORN
import seaborn as sns
sns.load_dataset('iris')
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
# independent & dependent feature
X = df.iloc[ : , :-1]
Y = dataset.target
```

```python
X,Y
```

FEATURE SCALING NEED NOT TO BE DONE HERE AS TIME COMPLEXITY MAY DECREASE SO NO USE OF MAKING OF DEDCISION TREE


```python
# Train Test Split 
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.33,random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# pOST Pruning
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train,Y_train)
```

```python
X_train.head()
```

```python
# Gini IMPURITY Is used by deafault 
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
# Post pruning
treeclassifier = DecisionTreeClassifier(criterion ='entropy',max_depth = 3)
treeclassifier.fit(X_train,Y_train)
```

```python
# entropy is used here 
from sklearn import tree
tree.plot_tree(treeclassifier,filled=True)
```

```python
## prediction 
Y_pred=treeclassifier.predict(X_test)
```

```python
Y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(Y_pred,Y_test)
print(score)
```

```python
classification_report(Y_pred,Y_test)
```
